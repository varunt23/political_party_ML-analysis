```python
import pandas as pd
import re
BSAfter = pd.read_csv("bs2cleanA.csv")
```

```python
BSAfter.head()
```

```python
def cleanTweet(tweet):
    tweet = re.sub(r'https?:\/\/.*\/\w*', '', str(tweet))
    tweet = re.sub('https','', str(tweet))
    return tweet

BSAfter['CleanText'] = BSAfter['CleanText'].apply(cleanTweet)
```

```python
BS_String1 = []
for t in BSAfter.CleanText:
    BS_String1.append(t)
BS_String1 = pd.Series(BS_String1).str.cat(sep = '')

import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=1600, height=800,max_font_size=200).generate(BS_String1)
plt.figure(figsize=(12,10)) 
plt.imshow(wordcloud, interpolation="bilinear") 
plt.axis("off") 
plt.show()

#colormap='magma'
```
